## Initial Dataframe Creation

```python
# Store the Json response to a varible
school_data = response["results"]
# Initialize the dataframe containing all the fields
school_data_df = pd.DataFrame(columns = ['open_admissions_policy', 'ownership_peps', 
                                         'price_calculator_url', 'city', 'ft_faculty_rate', 
                                         'locale', 'accreditor', 'main_campus', 'alias', 
                                         'faculty_salary', 'state', 'operating', 
                                         'instructional_expenditure_per_fte', 
                                         'state_fips', 'zip', 'carnegie_undergrad', 
                                         'women_only', 'religious_affiliation', 
                                         'tuition_revenue_per_fte', 'under_investigation', 
                                         'region_id', 'carnegie_basic', 'branches', 'men_only', 
                                         'school_url', 'online_only', 'ownership', 'accreditor_code', 
                                         'name', 'degree_urbanization', 'carnegie_size_setting', 
                                         'degrees_awarded_predominant', 'degrees_awarded_highest', 
                                         'institutional_characteristics_level', 'title_iv_approval_date', 
                                         'title_iv_eligibility_type', 'id', 'ope8_id', 'ope6_id',
                                         'location_lon', 'location_lat' 'tuition_in_state', 
                                         'tuition_out_of_state'])

# Create all the lists to store the individual queries
school_open_admissions_policy = [] # Open admissions policy indicator
school_ownership_peps = [] # Control of institution, per PEPS
school_price_calculator_url = [] # URL for institution's net price calculator
school_city = [] # City
school_ft_faculty_rate = [] # Proportion of faculty that is full-time
school_locale = [] # Locale of institution
school_accreditor = [] # Accreditor for institution
school_main_campus = [] # Flag for main campus
school_alias = [] # Institution name aliases
school_faculty_salary = [] # Average faculty salary
school_state = [] # State postcode
school_operating = [] # Flag for currently operating institution, 0=closed, 1=operating
school_instructional_expenditure_per_fte = [] # Instructional expenditures per full-time equivalent student
school_state_fips = [] # FIPS code for state
school_zip = [] # ZIP code
school_carnegie_undergrad = [] # Carnegie Classification -- undergraduate profile
school_women_only = [] # Flag for women-only college
school_religious_affiliation = [] # Religous affiliation of the institution
school_tuition_revenue_per_fte = [] # Net tuition revenue per full-time equivalent student
school_under_investigation = [] # Schools that are on Heightened Cash Monitoring 2 by the Department of Education
school_region_id = [] # Region (IPEDS)
school_carnegie_basic = [] # Carnegie Classification -- basic
school_branches = [] # Number of branch campuses
school_men_only = [] # Flag for men-only college
school_school_url = [] # URL for institution's homepage
school_online_only = [] # Flag for distance-education-only education
school_ownership = [] # Control of institution
school_accreditor_code = [] # Code corresponding to accreditor (as captured from PEPS)
school_name = [] # Institution name
school_degree_urbanization = [] # Degree of urbanization of institution
school_carnegie_size_setting = [] # Carnegie Classification -- size and setting

school_degrees_awarded_predominant = [] # Predominant undergraduate degree awarded
school_degrees_awarded_highest = [] # Highest degree awarded
school_institutional_characteristics_level = [] # Level of institution
school_title_iv_approval_date = [] # Date that institution was first approved to participate in Title IV aid programs
school_title_iv_eligibility_type = [] # Title IV eligibility type

school_id = [] # Unit ID for institution
school_ope8_id = [] # 8-digit OPE ID for institution
school_ope6_id = [] # 6-digit OPE ID for institution

school_location_lon = [] # Longitude
school_location_lat = [] # Latitude

school_tuition_in_state = [] # In-state tuition and fees
school_tuition_out_of_state = [] # Out-of-state tuition and fees

# Loop through the Json file to append the data to the lists 
for data in school_data:
    school_open_admissions_policy.append(data['school.open_admissions_policy'])
    school_ownership_peps.append(data['school.ownership_peps'])
    school_price_calculator_url.append(data['school.price_calculator_url'])
    school_city.append(data['school.city'])
    school_ft_faculty_rate.append(data['school.ft_faculty_rate'])
    school_locale.append(data['school.locale'])
    school_accreditor.append(data['school.accreditor'])
    school_main_campus.append(data['school.main_campus'])
    school_alias.append(data['school.alias'])
    school_faculty_salary.append(data['school.faculty_salary'])
    school_state.append(data['school.state'])
    school_operating.append(data['school.operating'])
    school_instructional_expenditure_per_fte.append(data['school.instructional_expenditure_per_fte'])
    school_state_fips.append(data['school.state_fips'])
    school_zip.append(data['school.zip'])
    school_carnegie_undergrad.append(data['school.carnegie_undergrad'])
    school_women_only.append(data['school.women_only'])
    school_religious_affiliation.append(data['school.religious_affiliation'])
    school_tuition_revenue_per_fte.append(data['school.tuition_revenue_per_fte'])
    school_under_investigation.append(data['school.under_investigation'])
    school_region_id.append(data['school.region_id'])
    school_carnegie_basic.append(data['school.carnegie_basic'])
    school_branches.append(data['school.branches'])
    school_men_only.append(data['school.men_only'])
    school_school_url.append(data['school.school_url'])
    school_online_only.append(data['school.online_only'])
    school_ownership.append(data['school.ownership'])
    school_accreditor_code.append(data['school.accreditor_code'])
    school_name.append(data['school.name'])
    school_degree_urbanization.append(data['school.degree_urbanization'])
    school_carnegie_size_setting.append(data['school.carnegie_size_setting'])
    school_degrees_awarded_predominant.append(data['school.degrees_awarded.predominant'])
    school_degrees_awarded_highest.append(data['school.degrees_awarded.highest'])
    school_institutional_characteristics_level.append(data['school.institutional_characteristics.level'])
    school_title_iv_approval_date.append(data['school.title_iv.approval_date'])
    school_title_iv_eligibility_type.append(data['school.title_iv.eligibility_type'])
    school_id.append(data['id'])
    school_ope8_id.append(data['ope8_id'])
    school_ope6_id.append(data['ope6_id'])
    school_location_lon.append(data['location.lon'])
    school_location_lat.append(data['location.lat'])
    school_tuition_in_state.append(data['2017.cost.tuition.in_state'])
    school_tuition_out_of_state.append(data['2017.cost.tuition.out_of_state'])
    
# Update the dataframe with the information in the lists    
school_data_df['open_admissions_policy'] = school_open_admissions_policy
school_data_df['ownership_peps'] = school_ownership_peps
school_data_df['price_calculator_url'] = school_price_calculator_url
school_data_df['city'] = school_city
school_data_df['ft_faculty_rate'] = school_ft_faculty_rate
school_data_df['locale'] = school_locale
school_data_df['accreditor'] = school_accreditor
school_data_df['main_campus'] = school_main_campus
school_data_df['alias'] = school_alias
school_data_df['faculty_salary'] = school_faculty_salary
school_data_df['state'] = school_state
school_data_df['operating'] = school_operating
school_data_df['instructional_expenditure_per_fte'] = school_instructional_expenditure_per_fte
school_data_df['state_fips'] = school_state_fips
school_data_df['zip'] = school_zip
school_data_df['carnegie_undergrad'] = school_carnegie_undergrad
school_data_df['women_only'] = school_women_only
school_data_df['religious_affiliation'] = school_religious_affiliation
school_data_df['tuition_revenue_per_fte'] = school_tuition_revenue_per_fte
school_data_df['under_investigation'] = school_under_investigation
school_data_df['region_id'] = school_region_id
school_data_df['carnegie_basic'] = school_carnegie_basic
school_data_df['branches'] = school_branches
school_data_df['men_only'] = school_men_only
school_data_df['school_url'] = school_school_url
school_data_df['online_only'] = school_online_only
school_data_df['ownership'] = school_ownership
school_data_df['accreditor_code'] = school_accreditor_code
school_data_df['name'] = school_name
school_data_df['degree_urbanization'] = school_degree_urbanization
school_data_df['carnegie_size_setting'] = school_carnegie_size_setting

school_data_df['degrees_awarded_predominant'] = school_degrees_awarded_predominant
school_data_df['degrees_awarded_highest'] = school_degrees_awarded_highest
school_data_df['institutional_characteristics_level'] = school_institutional_characteristics_level
school_data_df['title_iv_approval_date'] = school_title_iv_approval_date
school_data_df['title_iv_eligibility_type'] = school_title_iv_eligibility_type

school_data_df['id'] = school_id
school_data_df['ope8_id'] = school_ope8_id
school_data_df['ope6_id'] = school_ope6_id

school_data_df['location_lon'] = school_location_lon
school_data_df['location_lat'] = school_location_lat

school_data_df['tuition_in_state'] = school_tuition_in_state
school_data_df['tuition_out_of_state'] = school_tuition_out_of_state



# Print the final dataframe
school_data_df.head()
```
